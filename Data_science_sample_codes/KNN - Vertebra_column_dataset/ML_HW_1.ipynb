Segregated all abnormal instances into a seperate dataframe called abnormality_class and all normal instances into a seperate dataframe called the normal_class and then performed the split as mentioned in the question by considering first 70 rows of normal class and first 140 rows of the abnormal class and concatenating them to generate the required training data . Thereafter split the total training data into final_training_data_x (containing the predictors of the training data) and final_training_data_y (containing the response of the training data).Thereafter shuffle both of them to change the order of training instances to eliminate any potential bias due to any form of order in the training data .

```
from sklearn.utils import shuffle

abnormality_class = df_final.loc[df_final['class']== 1]
normal_class = df_final.loc[df_final['class']==0]
train_set_class1 = abnormality_class.iloc[0:140,:]
train_set_class0 = normal_class.iloc[0:70,:]

training_list=[train_set_class0,train_set_class1]
final_training_data = pd.concat(training_list)

final_training_data_x = final_training_data.iloc[0:210,0:6]
final_training_data_y = final_training_data.iloc[0:210,6:7]

final_training_data_x,final_training_data_y = shuffle(final_training_data_x,final_training_data_y)
```

```
print(final_training_data_x.shape)
print(final_training_data_y.shape) 
```

Thereafter split the total test data into final_testing_data_x (containing the predictors of the test data) and final_testing_data_y (containing the response of the test data) and shuffle both of them to generate a random seed of test data points.

# c) 3. Best test error rate vs size of training set {learning curve} :

Ran a for loop for N being in a range of values as given in the question and then for each iteration of N ; I alloted floor(N/3) of class 0 instances and N - floor(N/3) class 1 instances for my training data during that iteration . Thereafter I ran a nested loop for the KNN algorithm with K being in a range of values and then generated the test error rate list from which I selected the least value . I repeated the same process for N iterations and generated the same no of minimal test errors as the no of times the outer loop ran . Thereafter I plotted the minimal test error rate from each one of those iterations along Y axis and the training set size during each of those iterations along X -axis and got my learning curve.

```
import math
from sklearn.utils import shuffle
N = list(range(10,210,10))


best_test_er = []

for q in N :
  flr_class0 = math.floor(q/3)
  flr_class1 = q - math.floor(q/3)
  sum_flr = flr_class0 + flr_class1
  train_subset_class1 = abnormality_class.iloc[0:flr_class1,:]
  train_subset_class0 = normal_class.iloc[0:flr_class0,:]
  
  train_subset_combo = [train_subset_class0,train_subset_class1]
  final_training_subset_data = pd.concat(train_subset_combo)
  
  final_training_subset_data_x = final_training_subset_data.iloc[0:sum_flr,0:6]
  final_training_subset_data_y = final_training_subset_data.iloc[0:sum_flr,6:7]
  
  
  final_training_subset_data_x,final_training_subset_data_y = shuffle(final_training_subset_data_x,final_training_subset_data_y)
  k_values_lc = list(range(1,196,5))

  scores_lc= {}
  sc_list_lc=[]
  training_score_lc=[]
  testing_score_lc=[]  
  
  
  for u in k_values_lc :
    mdl_lc=KNeighborsClassifier(n_neighbors=u)
    mdl_lc.fit(final_training_subset_data_x, final_training_subset_data_y)
    if(u <= len(final_training_subset_data_x) ): 
    
      predicted_val_lc = mdl_lc.predict(final_testing_data_x)
      scores_lc[u]=metrics.accuracy_score(final_testing_data_y,predicted_val_lc)
      sc_list_lc.append(metrics.accuracy_score(final_testing_data_y,predicted_val_lc))
      training_score_lc.append(mdl_lc.score(final_training_subset_data_x,final_training_subset_data_y))
      testing_score_lc.append(mdl_lc.score(final_testing_data_x,final_testing_data_y))
    else :
      break
      
  unity_lc = [1] *len(testing_score_lc)
  test_error_rate_lc = np.subtract(unity_lc,testing_score_lc)
  
  best_test_er.append(min(test_error_rate_lc))
  
  
```

**Learning curve plotted**

The best test error rate drops and then rises up a bit before it becomes constant for a few iterations of N after which it drops continuously for further iterations of N .

```
print(best_test_er)
import matplotlib.pyplot as plt1
plt1.plot(N,best_test_er,c='green')                               #learning curve plotted below
plt1.xlabel('Training set size')
plt1.ylabel('best test error rate')
```
